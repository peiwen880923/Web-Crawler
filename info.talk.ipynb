```python
#統整前面的程式碼，擷取標題、副標題、作者、時間、內文
import requests
from bs4 import BeautifulSoup

url='https://info.talk.tw/'
r=requests.get(url)
resp=BeautifulSoup(r.text,'lxml')
all_content=resp.find_all('div','themeA post-grid-panel panel')[0]
content=all_content.find_all('div','entry-content')
items=[]
for i in content:
  title=i.find('h3','entry-title').text.replace('\n','').replace('\r','').replace(' ','')
  subtitle=i.find('p','txt').text.replace('\n','').replace('\r','').replace(' ','')
  time=i.find('span','item date').text.replace('\n','').replace('\r','').replace(' ','')
  author=i.find('span','item writer').text.replace('\n','').replace('\r','').replace(' ','')
  link="https://info.talk.tw/"+i.find('h3','entry-title').find('a').get('href').replace(' ','')
  r=requests.get(link)
  resp=BeautifulSoup(r.text,'lxml')
  article=resp.find('div','article-content').text.replace('\n','').replace('\r','').replace(' ','')
  items.append([title,subtitle,time,author,link,article])
print(items)
```

```python
#將擷取的內容存成csv
import csv
with open('comment.csv','w',encoding='utf-8',newline='') as f: #開啟輸出的CSV檔案
  writer=csv.writer(f) #建立CSV檔寫入器
  writer.writerow(['標題','副標題','時間','作者','連結','內文']) #csv的標題
  for item in items:
    writer.writerow(item)
```
