```python
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions
from selenium.webdriver.common.by import By
url="https://www.booking.com/reviews/tw/hotel/yang-yang-wei-lu-jezreel-inn.zh-tw.html?label=gen173nr-1FCFQoggJCCWthb2hzaXVuZ0gwWARo5wGIAQGYATC4ARfIAQzYAQHoAQH4AQKIAgGoAgO4AtvcqpoGwAIB0gIkZmFjZDMyMTYtNjY1Ny00ODA3LTg5OTAtZDlhODMwZmVmZTk02AIF4AIB&sid=071295ad96b16ef0b4793c9d6435172e&customer_type=total&hp_nav=0&old_page=0&order=featuredreviews&page=1&r_lang=zh-tw&rows=75&"
items=[]
while True:
    options = webdriver.ChromeOptions()
    driver=webdriver.Chrome(chrome_options=options,executable_path="chromedriver")
    driver.set_page_load_timeout(60)
    driver.get(url)
    soup = BeautifulSoup(driver.page_source, 'lxml')
    all_1=soup.find_all('div','review_item_review')
    for i in all_1:
        items1=[]
        try:
            score=i.find("span","review-score-badge").text.replace('\n','').replace('\r','')
            items1.append(score)
            title=i.find('div','review_item_header_content_container').text.replace('\n','').replace('\r','')
            items1.append(title)
            positive=i.find('p','review_pos').find('span',itemprop="reviewBody").text.replace('\n','').replace('\r','')
            items1.append(positive)
        except:
            positive="none"
            items1.append(positive)
        try:
            negative=i.find('p','review_neg').find('span',itemprop="reviewBody").text.replace('\n','').replace('\r','')
            items1.append(negative)
        except:
            negative="none"
            items1.append(negative)
        items.append(items1)
    try:
        #print(soup.find('p','page_link review_next_page').find('a').get("href"))
        url="https://www.booking.com/"+soup.find('p','page_link review_next_page').find('a').get("href")
        print(url)
    except:
        print("最後一頁!!!")
        break
        
driver.quit()
print(items)
```
